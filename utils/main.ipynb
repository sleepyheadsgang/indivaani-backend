```python
from translator import translate
```

```python
def is_valid_sentence(sentence):
    sentence = str(sentence)
    sentence = sentence.replace(' ', '')
    if len(sentence.strip()) == 1:
        return False
    sentence = sentence.replace('-', '')
    sentence = sentence.replace('_', '')
    # print(sentence.isalnum(), type(sentence),
    #   "My sentence", sentence, sep="--")
    # print(sentence)
    return sentence.isalpha() or sentence.isalnum()
```

```python
def iterate_nodes(node):
    for i in node.children:
        if isinstance(i, bs4.element.NavigableString):
            if i.strip() != "":
                # print("Valid sentence", i)
                if is_valid_sentence(i):
                    i.replace_with(translate(i, "en", "hi"))
                    # print(i)
        else:
            iterate_nodes(i)

iterate_nodes(list(body_content.children)[3])
```
